```python
def checkClones(group, n):
    checker = True
    for i in range(n):
        if group[i][1] != sourcedir:
            checker = False
        if group[i][4] == skip_method_name:
            checker = False
    for i in range(n-1):
        if group[i][2] != group[i+1][2]:
            checker = False
    return checker        
            
def isRepeatedGroup(group, groupSet, n):
    methods = []
    for i in range(n):
        clone = group[i]
        method = clone[2] + '.' + clone[3] + '#' + clone[4] + '@' + str(clone[5])
        methods.append(method)
    for method in methods:
        if method in groupSet:
            return True
    for method in methods:
        groupSet.add(method)
    return False 
    
```
