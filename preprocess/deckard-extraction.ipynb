```python
import re

method_name_pattern = re.compile(r'^(.*)\s(.*?)[\s]*\(')
path_pattern = re.compile(r'/home/yannick/workspace/eclipse/workspace/jfreechart-1.0.10-rf/(.*?)/(.*)/(.*?).java')

def getMethodName(path, startLine, kind):
    if kind == 113:
        javaFile = open(path, 'r')
        line = javaFile.readlines()[int(startLine) - 1]
        javaFile.close()
        m = method_name_pattern.match(line)
        return m.group(2)

def getCUInfo(path):
    m = path_pattern.match(path)
    return m.group(1), m.group(2).replace('/', '.'), m.group(3)
```

```python
fileName = 'data/post_cluster_vdb_50_0_allg_0.95_50'
fin = open(fileName, 'r')
records = fin.readlines()
fin.close()

for record in records:
    tokens = record.split()
    if len(tokens) > 0:
        path = tokens[3]
        folder, package, className = getCUInfo(path)
        [startLine, length] = tokens[4].split(':')[1:3]
        kind = int(tokens[5].split(':')[1])
        startLine = int(startLine)
        length = int(length)
        method = getMethodName(path, startLine, kind)
        if method != None:
            print(folder, package, className, startLine, length, kind, method)
```
